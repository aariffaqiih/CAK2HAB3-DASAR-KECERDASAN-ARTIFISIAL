### 8. Penyalinan (Copy) Array (`5`)

**a.** Seorang ilmuwan data ingin melakukan analisis terhadap data eksperimen dengan mengubah data aslinya. Buat salinan "shallow copy" dari array data asli, lalu ubah elemen tertentu pada indeks tertentu dari data yang telah disalin.

```python
# Data eksperimen asli
data_asli = np.array([10, 15, 20, 25, 30])

# Buat shallow copy
data_copy = data_asli

# Masukkan indeks dan data baru
informasi_perubahan = input("masukkan index yang ingin diubah dan data baru: ").split(",")

# Ubah data
data_copy[int(informasi_perubahan[0])] = int(informasi_perubahan[1])

# Tampilkan hasil
print("data copy:", data_copy)
print("data asli:", data_asli)
```

**b.** Seorang peneliti ingin melakukan eksperimen dengan data cuaca, dan memastikan bahwa perubahan data baru tidak mempengaruhi data aslinya. Buat "deep copy" dari array cuaca dan ubah data salinan tersebut.

```python
# Data cuaca asli
data_cuaca = np.array([[30, 32, 35], [28, 27, 26], [33, 31, 29]])

# Buat deep copy dari data_cuaca
data_cuaca_copy = data_cuaca.copy()

# Ubah elemen di baris ke-2, kolom ke-3 data_cuaca_copy menjadi 40
data_cuaca_copy[1, 2] = 40

# Tampilkan data sesuai contoh output
print("data cuaca copy:\n", data_cuaca_copy)
print("data cuaca asli:\n", data_cuaca)
```

### 9. Operasi Dasar Array (`5`)

**a.** Seorang guru ingin mengetahui nilai tertinggi yang diperoleh siswa dari hasil ujian matematika. </br> Tambahkan juga kode untuk menghitung selisih antara nilai maksimum dan nilai rata-rata ujian.

```python
# Data nilai ujian siswa dalam array
nilai_ujian = np.array([75, 80, 90, 85, 60])

# Temukan nilai maksimum
nilai_maksimum = np.max(nilai_ujian)

# Hitung nilai rata-rata
nilai_rata_rata = np.mean(nilai_ujian)

# Hitung selisih antara nilai maksimum dan rata-rata
nilai_selisih = nilai_maksimum - nilai_rata_rata

# Tampilkan nilai maksimum dan selisihnya
print(f"Nilai Maksimum: {nilai_maksimum}")
print(f"Selisih Maksimum dan Rata-rata: {nilai_selisih}")
```

```python
# Data waktu renang atlet
waktu_renang = np.array([50.5, 52.0, 48.7, 49.8, 51.2])

# Waktu tercepat
waktu_tercepat = np.min(waktu_renang)

# Rata-rata
waktu_rata_rata = np.mean(waktu_renang)

# Selisih (dibulatkan 1 angka desimal)
selisih = round(waktu_tercepat - waktu_rata_rata, 0)

print(f"Waktu Tercepat: {waktu_tercepat}")
print(f"Selisih Waktu Tercepat dan Rata-rata: {selisih}")
```

### 11. Soal Cerita (`20`)

Seorang manajer toko retail sedang menganalisis tingkat kepuasan pelanggan (skala 5 - 10) dari tiga cabang toko selama 8 minggu. Data kepuasan disimpan dalam sebuah array yang merepresentasikan rating kepuasan dari tiga cabang (kolom) di setiap minggu (baris). Manajer hanya tertarik pada minggu-minggu di mana rata-rata kepuasan ketiga cabang lebih dari atau sama dengan 8. Mereka ingin menganalisis lebih lanjut dengan menyalin data minggu tersebut, melakukan transpose pada data untuk melihat statistik dari perspektif cabang, dan menghitung rata-rata kepuasan tiap cabang. Berikut adalah langkah-langkah yang perlu dilakukan:

- Ambil subset data dari minggu-minggu yang dimana rata-rata kepuasan 3 cabang pada minggu yang sama memiliki nilai lebih dari atau sama dengan 8
- Buat salinan data untuk dianalisis lebih lanjut
- *Transpose array* hasil salinan statistik
- Ambil rata-rata kepuasan tiap cabang pada minggu (baris) yang terdapat pada subset dari data kepuasan

```python
# Buat array yang merepresentasikan data kepuasan 8 minggu dari 3 cabang (8x3)
data_kepuasan = np.array([
    [9, 7, 8],
    [8, 9, 10],
    [7, 6, 8],
    [10, 8, 7],
    [6, 7, 5],
    [8, 8, 9],
    [5, 6, 7],
    [9, 10, 8]
])

# You're on Your Own, Kid. Good Luck :)

# Tampilkan data asli
print("Data Kepuasan Asli:")
print(data_kepuasan)

# Hitung rata-rata tiap minggu
rata_mingguan = np.mean(data_kepuasan, axis=1)

# Ambil minggu dengan rata-rata >= 8
subset = data_kepuasan[rata_mingguan >= 8]

print("Subset Data Kepuasan:")
print(subset)

# Buat salinan data
subset_copy = subset.copy()

# Transpose
transpose_data = subset_copy.T

print("Data Setelah Di-transpose:")
print(transpose_data)

# Hitung rata-rata tiap cabang
rata_cabang = np.mean(transpose_data, axis=1)

print("Rata - Rata Kepuasan Cabang A: ", rata_cabang[0])
print("Rata - Rata Kepuasan Cabang B: ", rata_cabang[1])
print("Rata - Rata Kepuasan Cabang C: ", rata_cabang[2])
```
