### Import *Library* NetworkX dan Matplotlib

*Library* Matplotlib digunakan untuk membantu menampilkan visualisasi graf pada terminal.

````python
import networkx as nx # Import library NetworkX
import matplotlib.pyplot as plt # Import library Matplotlib
````

### Contoh fungsi pendukung untuk mencetak graf

#### **!! Tidak perlu dimodifikasi !!**

````python
# Fungsi pendukung untuk mencetak graf
def show_graph(G, pos=None, title='') :
    # Jika posisi tidak ditentukan dari parameter maka akan generate posisi baru
    if pos == None :
        # Variabel pos yang menyimpat koordinat penempatan node
        pos = nx.spring_layout(G)

    # Plot graf G yang diberikan pada parameter
    nx.draw(G, pos, with_labels=True, node_color='red', node_size=2500,
            font_color="white", font_weight="bold", width=5)

    # Mengambil setiap weight dari setiap edge pada graf G
    edge_labels = nx.get_edge_attributes(G, "weight")

    # Menggambar weight dari edge pada visualisasi
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='blue',
                                font_weight="bold", font_size=12)

    # Menambahkan margin pada setiap sisi
    plt.margins(0.2)
    # Menambahkan judul sesuai yang diberikan oleh parameter
    plt.title(title)
    # Menampilkan visualisasi graf pada terminal
    plt.show()
````

**c. (10 poin)** Pada graf town, setiap edge mewakili hubungan antara dua lokasi, dan masing-masing hubungan memiliki jarak tempuh tertentu yang direpresentasikan sebagai `weight` pada graf. Berikut ini adalah daftar jarak antar lokasi yang ada di kota tersebut:

- `Balai Kota (Balai)` ke `Bandara` memiliki jarak `5km`.
- `Balai Kota (Balai)` ke `Pelabuhan` memiliki jarak `4km`.
- `Pelabuhan` ke `Kampus` memiliki jarak `6km`.
- `Bandara` ke `Mall` memiliki jarak `2km`.
- `Mall` ke `Pelabuhan` memiliki jarak `2km`.
- `Bandara` ke `Pelabuhan` memiliki jarak `7km`.

Tugas Anda adalah menambahkan edge ke dalam graf untuk setiap hubungan lokasi yang terdaftar di atas. Pastikan setiap edge mencakup informasi jarak tempuh (weight) sebagai atribut tambahan.

````python
# Daftar hubungan (edge) antar lokasi di Mekarjaya Town beserta jarak tempuh (weight) masing-masing
edges = [
    ('Balai', 'Bandara', {'weight': 5}),
    ('Balai', 'Pelabuhan', {'weight': 4}),
    ('Pelabuhan', 'Kampus', {'weight': 6}),
    ('Bandara', 'Mall', {'weight': 2}),
    ('Mall', 'Pelabuhan', {'weight': 2}),
    ('Bandara', 'Pelabuhan', {'weight': 7})
]

# Tambahkan edge beserta weightnya ke dalam graf town
town.add_edges_from(edges)

# Tampilkan graf town setelah penambahan edge
show_graph(town, pos, title="Menambahkan Edge pada Graf")

# di sini kita menggunakan list edges yang sudah dibuat sebelumnya, sehingga semua hubungan antar lokasi beserta jaraknya akan otomatis masuk ke graf.
# show_graph() akan menggambar graf berdasarkan posisi pos yang sudah ditentukan,
# dan menampilkan title "Menambahkan Edge pada Graf" agar user tahu bahwa graf ini menampilkan hubungan antar node yang baru saja ditambahkan.
# visualisasi ini berguna untuk melihat bagaimana node-node saling terhubung dan jarak antar lokasi di dalam graf.
````

**e. (10 poin)** Buatlah sebuah program yang meminta input dari pengguna untuk menentukan:
- **Node awal** (`start_node`): Titik awal perjalanan Riko di kota Mekarjaya Town.
- **Node tujuan** (`end_node`): Titik tujuan yang ingin dicapai oleh Riko.

💡 Clue:
- Gunakan `nx.shortest_path(G, source, target, weight='weight')` untuk mendapatkan urutan lokasi yang harus dilewati dari **node awal ke node tujuan** berdasarkan jarak tempuh terpendek.
- Gunakan `nx.shortest_path_length(G, source, target, weight='weight')` untuk mendapatkan **total jarak tempuh** dari **node awal ke node tujuan**.

````python
def jalan_mekarjaya():
    # Input titik awal dan tujuan
    start_node = input("Masukkan titik awal: ").title()
    end_node = input("Masukkan titik tujuan: ").title()

    try:
        # Hitung jalur terpendek dan jaraknya
        shortest_path = nx.shortest_path(town, source=start_node, target=end_node, weight='weight')
        shortest_distance = nx.shortest_path_length(town, source=start_node, target=end_node, weight='weight')

        # Menampilkan hasil
        print(f"Jalur terpendek dari {start_node} ke {end_node}: {shortest_path}")
        print(f"Total jarak tempuh: {shortest_distance} km")

    except nx.NetworkXNoPath:
        print("Tidak ada jalur yang menghubungkan kedua lokasi tersebut.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

# program ini berfungsi untuk meminta user memasukkan titik awal dan titik tujuan perjalanan di kota Mekarjaya. pertama user diminta input titik awal,
# kemudian titik tujuan. .title() digunakan agar input tetap bisa diterima walaupun user mengetik dengan huruf kecil atau besar campur-campur,
# sehingga format penulisan nama lokasi menjadi konsisten sesuai node di graf.
# pertama kita bungkus dengan try untuk jaga-jaga kalau terjadi error misalnya input lokasi tidak ada di graf atau tidak ada jalur.
# kemudian nx.shortest_path() digunakan untuk mendapatkan urutan node yang dilewati dari start ke end dengan mempertimbangkan atribut weight sebagai jarak.
# sedangkan nx.shortest_path_length() digunakan untuk menghitung total jarak tempuhnya berdasarkan weight yang sudah ada di graf.
# jika tidak ada jalur yang menghubungkan kedua lokasi maka akan muncul pesan "Tidak ada jalur yang menghubungkan kedua lokasi tersebut."
# sedangkan jika terjadi error lain misalnya input salah atau graf belum dibuat dengan benar,
# maka akan ditangani oleh except Exception dan menampilkan pesan error yang sesuai.
# ini berguna supaya program tetap aman dan tidak berhenti tiba-tiba saat user memasukkan data yang tidak valid.
````

````python
jalan_mekarjaya()

# menjalankan prosedur jalan_mekarjaya yang sebelumnya sudah dibuat untuk menghitung jalur terpendek antara dua lokasi di kota Mekarjaya.
# tugas kita adalah memanggil prosedur ini, lalu prosedur akan meminta user input titik awal dan titik tujuan, menghitung jalur terpendek beserta jaraknya.
````
