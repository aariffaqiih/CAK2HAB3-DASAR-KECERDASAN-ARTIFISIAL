**b.** Buat program untuk meminta *input* usia dari pengguna. Program tersebut menerima usia dan menampilkan kategori usia pada terminal sesuai dengan kriteria berikut
- **Balita**: 0 - 5
- **Anak-anak**: 6 - 12

**b.** Buatlah sebuah **fungsi** sederhana yang mengonversi suhu dari Celsius (C) menjadi Fahrenheit (F). Fungsi ini akan menerima input berupa suhu dalam Celsius, kemudian mengembalikan hasil konversi dalam Fahrenheit.

Rumus: $F = (C \times \frac{9}{5}) + 32$

````python
def celsius_to_fahrenheit(c):
  return (c * 9/5) + 32

try:
  c = float(input("Masukkan suhu (C): "))
except ValueError:
  print("Masukan suhu tidak valid.")
else:
  f = celsius_to_fahrenheit(c)
  print(f"Dalam fahrenheit = {f}")
````

**b.** Buatlah sebuah fungsi yang menerima argumen `n` dan mengembalikan hasil perkalian (faktorial) dari bilangan bulat dari 1 hingga n. </br>Sebagai contoh, jika `n = 5`, fungsi tersebut harus menghitung perkalian **$1 \times 2 \times 3 \times 4 \times 5$** dan mengembalikan nilai 120. </br>Jika `n` kurang dari 1, kembalikan nilai 1. Gunakan perulangan `while` untuk menghitung hasil tersebut.

````python
def faktorial(n): # Buat fungsi untuk menghitung faktorial
  if n < 1:       # Kondisi jika n kurang dari 1
    return 1
  else:
    hasil = 1
    i = 1
    while i <= n: # Buat kondisi untuk melakukan perulangan dari 1 hingga n
      hasil *= i  # Kalikan hasil dengan i
      i += 1      # Tambah i dengan 1 sebagai iterator

    return hasil  # Mengembalikan variabel yang menampung nilai hasil

try:
  # Buat kode untuk meminta input pengguna dan konversi input ke integer
  user_input = int(input("Masukkan nilai n: "))
  # Panggil fungsi untuk menghitung faktorial dan simpan hasilnya kedalam variabel
  # Tampilkan hasil faktorial yang sudah dihitung ke terminal
  print(f"Hasil faktorial dari {user_input} adalah {faktorial(user_input)}")
except ValueError as e:
  print("Input tidak valid:", e)
````

**a.** Buat program yang mengganti status `"pending"` menjadi `"completed"` pada setiap elemen dalam list `tasks`. Gunakan metode `.replace()` untuk mengubah status, kemudian simpan status yang telah diubah ke dalam list baru bernama `updated_tasks` dan kemudian cetak list ke dalam terminal.

Tip :

````python
tasks = ["task1: pending", "task2: pending", "task3: pending", "task4: pending"]
updated_tasks = [] # Buat list kosong

for task in tasks : # Buat loop untuk iterasi elemen pada list tasks
    new_status = task.replace("pending", "completed") # Ganti status pending menjadi completed menggunakan .replace()
    updated_tasks.append(new_status) # Masukkan status yang sudah diganti ke dalam list updated_tasks menggunakan .append()

# Cetak list yang berisi status yang sudah diganti
print("Status baru:", updated_tasks)
````

**b.** Buatlah fungsi `nomor_telepon` yang menerima sebuah dictionary. Dictionary ini berisi kode negara sebagai `key`, dan daftar nomor lokal sebagai `value`. Lengkapi bagian yang kosong dalam fungsi untuk menghasilkan daftar yang berisi nomor telepon lengkap. Nomor telepon ini dihasilkan dengan menggabungkan setiap nomor lokal dengan kode negara yang sesuai.

Tip:
```python
hobi_dict = {
    "Budi" : "Membaca",
    "Siti" : "Menulis",
    "Andi" : "Bermain Game"
}

for key, value in hobi_dict.items() :
    print("Key:", key, ", Value:", value)
```

Output
```
Key: Budi , Value: Membaca
Key: Siti , Value: Menulis
Key: Andi , Value: Bermain Game
```

````python
def nomor_telepon(kode_negara) :
    telepon_lengkap = [] # List untuk menampung nomor telepon lengkap

    # Loop untuk setiap key dan value pada dict kode_negara
    for kode, daftar_nomor  in kode_negara.items() :
        for nomor in daftar_nomor:
            # Tambahakan nomor telepon lengkap ke list telepon_lengkap
            telepon_lengkap.append(kode+'-'+nomor)

    return telepon_lengkap

data_telepon = {
    "+62": ["81234567890", "82198765432", "85312345678"],
    "+1" : ["2025551234", "3105559876"],
    "+44": ["7911123456", "7700900123"]
}

# Panggil fungsi nomor_telepon, kemudian masukkan ke dalam variabel
hasil = nomor_telepon(data_telepon)

# Cetak list hasil nomor telepon lengkap yang didapatkan dari nomor_telepon
print(hasil)
````

### 6. Soal Cerita (Bobot Nilai: `40`)

Seorang pemilik perpustakaan ingin membuat program sederhana untuk mencatat **peminjaman buku harian** oleh anggota. Setiap anggota akan dicatat jumlah buku yang dipinjam selama **tiga waktu** dalam sehari (pagi, siang, sore). Pemilik perpustakaan perlu menghitung **total buku yang dipinjam** untuk setiap anggota dan menentukan apakah anggota tersebut mendapat **diskon** atau **tidak**. Kriteria diskon adalah jika total buku yang dipinjam sama dengan atau lebih dari **10 buku**.

Buatlah program yang melakukan hal berikut :
1. **Input/Output**
    - Meminta input dari pengguna berupa `nama anggota` dan `jumlah buku yang dipinjam` untuk tiga waktu (pagi, siang, sore)
    - Menampilkan **total peminjaman** dan **status diskon** setiap anggota

2. **Penentuan diskon**
    - Menentukan apakah anggota **dapat diskon** atau **tidak dapat diskon** berdasarkan total buku yang dipinjam.

3. **Pengulangan program**
    - Menggunakan perulangan untuk memasukkan data beberapa anggota hingga pengguna memasukkan `"selesai"` sebagai nama anggota

4. **Perhitungan total**
    - Membuat fungsi untuk **menghitung total** yang menerima list peminjaman sebagai parameter dan mengembalikan total buku dari list tersebut

5. **Penyimpanan data**
    - Gunakan *dictionary* dengan `keys` nama anggota, dan `value` list peminjaman dari anggota tersebut.

````python
# fungsi untuk menjumlahkan buku yang dipinjam
def hitung_total(daftar_peminjaman):
  # sum = menjumlahkan
  return sum(daftar_peminjaman)

# fungsi untuk cek apakah inputnya int/valid
def read_int(book):
  while True:
    try:
      return int(input(book))
    except ValueError:
      print("Masukan tidak valid. Masukkan angka bulat.")

# fungsi utama program
def main():
  # dictionary untuk menyimpan data peminjam
  data_anggota = {}
  while True:
    # diberi strip untuk jaga2 siapa tau kelebihan spasi
    nama = input("Masukkan nama anggota (atau 'selesai' untuk berhenti): ").strip()
    if nama.lower() == "selesai":
      break
    pagi  = read_int("Masukkan jumlah buku pagi: ")
    siang = read_int("Masukkan jumlah buku siang: ")
    sore  = read_int("Masukkan jumlah buku sore: ")
    data_anggota[nama] = [pagi, siang, sore]

  print("\n=== Rekap Peminjaman Anggota ===")
  if not data_anggota:
    print("(tidak ada data anggota)")
    return

  for nama, peminjaman in data_anggota.items():
    total = hitung_total(peminjaman)
    # perhitungan apakah dapat diskon
    status = "Dapat Diskon" if total >= 10 else "Tidak Dapat Diskon"
    print(f"Nama: {nama} | Peminjaman: {peminjaman} | Total: {total} buku | Status: {status}")

if __name__ == "__main__":
  main()
````
